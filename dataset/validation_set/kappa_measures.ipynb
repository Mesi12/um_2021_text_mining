# Kappa measures

To generate kappa measures, a random sample of all our data instances has to be created. This notebook does that job.

```python
import os
import random
import re
import nltk
```

```python
with open("../original/novels/stud.txt") as fp:
    data = fp.read()

# text_after = re.sub(regex_search_term, regex_replacement, text_before)
data = data.replace("\n", " ")
data = re.sub(r" +", " ", data)
sent = nltk.tokenize.sent_tokenize(data)
```

```python
random.Random().shuffle(sent)

sent
```
